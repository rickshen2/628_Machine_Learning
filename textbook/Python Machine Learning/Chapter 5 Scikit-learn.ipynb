```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# from matplotlib import style
# import seaborn as sns

# style.use("ggplot")
heights = [[1.6],[1.65],[1.7],[1.73],[1.8]]
weights = [[60],[65],[72.3],[75],[80]]

lmod = LinearRegression()
lmod.fit(X=heights, y= weights)

plt.title("weight vs height")
plt.plot(heights, weights, 'k.')
plt.grid(True)

plt.plot(heights, lmod.predict(heights), color = 'r')

extreme_heights = [[0],[2.2]]
plt.plot(extreme_heights, lmod.predict(extreme_heights), color ='b')

# path = "C:\\Temp\\563_Data\\"
# filename = "B4.csv"
# df = pd.read_csv(path+filename)
# print(df.head(5))

# plt.scatter(df.y, df.x1)
# sns.pairplot(df)
# sns.get_dataset_names()
# sns.lmplot(data=df,x=df.x1, y=df.y)
```
