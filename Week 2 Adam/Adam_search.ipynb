```python
import torch
import torch.optim as optim
import matplotlib.pyplot as plt

# Define the quadratic function: f(x, y) = x^2 + y^2
def quadratic_function(params):
    x, y = params
    return x**2 + y**2

# Initialize the variables x and y with random starting points
params = torch.tensor([2.0, 3.0], requires_grad=True)  # Starting at (2, 3)

# Create an Adam optimizer
optimizer = optim.Adam([params], lr=0.1)  # Learning rate 0.1

# Store the history of parameter values and function values for plotting
x_history = []
y_history = []
loss_history = []

# Perform gradient descent for 100 iterations
for step in range(100):
    # Zero the gradients
    optimizer.zero_grad()

    # Compute the loss (the value of the quadratic function)
    loss = quadratic_function(params)

    # Perform backpropagation to compute the gradients
    loss.backward()

    # Update the parameters using the Adam optimizer
    optimizer.step()

    # Store the current values of x, y, and the function value
    x_history.append(params[0].item())
    y_history.append(params[1].item())
    loss_history.append(loss.item())

    # Print the current state every 10 steps
    if step % 10 == 0:
        print(f'Step {step}: x = {params[0].item()}, y = {params[1].item()}, f(x, y) = {loss.item()}')

# Plot the optimization process

# Plot x and y values over iterations
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(x_history, label='x values')
plt.plot(y_history, label='y values')
plt.title('Convergence of x and y')
plt.xlabel('Iteration')
plt.ylabel('Parameter value')
plt.legend()

# Plot the function values over iterations
plt.subplot(1, 2, 2)
plt.plot(loss_history)
plt.title('Convergence of f(x, y)')
plt.xlabel('Iteration')
plt.ylabel('f(x, y)')

plt.tight_layout()
plt.show()
```

```python
plt.plot(x_history, y_history)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Adam Optimization Process')
for i in range(len(x_history)):
  if i % 10  == 0:
    plt.scatter(x_history[i],y_history[i],color='red')
plt.show()
```
