```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 0.1, dtype=float)

silu = x / (1 + np.exp(-x))
relu = np.maximum(0, x)

plt.plot(x, silu, label='SiLU')
plt.plot(x, relu, label='ReLU')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('activation function')
plt.legend()  # Display the legend with the labels
plt.show()
```
